```python
import cv2
```

```python
img=cv2.imread("images..jfif")
print(img)
cv2.imshow("output",img)
cv2.waitKey(0)
```

```python
cap=cv2.VideoCapture("video.mp4")
while True:
    success ,img=cap.read()
    cv2.imshow("Output",img)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
```

```python
cap=cv2.VideoCapture(0)
cap.set(3,640)
cap.set(4,480)
while True:
    success ,img=cap.read()
    cv2.imshow("Output",img)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
```
